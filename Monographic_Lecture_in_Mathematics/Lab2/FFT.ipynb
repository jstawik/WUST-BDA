```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack as fft
```

```python
N = 1024
ts = np.linspace(0, 2, N)
signal = list(map(lambda x: np.sin(np.pi*100*x), ts))
plt.plot(signal)
```

```python
plt.plot(fft.fft(signal)[:N//2])
plt.show()

delta=ts[2]-ts[1]
xFFT=fft.fft(signal)
freqsFFT=np.linspace(0., 1./(2*delta), N//2)
plt.plot(freqsFFT, 2*np.abs(xFFT[:N//2]/N))
plt.show()
```

```python
A = np.sin(2*np.pi*25*ts) + np.sin(2*np.pi*50*ts)
```
