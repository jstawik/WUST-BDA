### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

### SIR model
$$
\begin{cases}
\frac{dS(t)}{dt} = -\frac{\beta S(t) I(t)}{N} \\
\frac{dI(t)}{dt} = \frac{\beta S(t) I(t)}{N}-\gamma I(t)\\
\frac{dR(t)}{dt} = \gamma I(t)
\end{cases}
$$

```python
N = 800.
I_0 = 1.
S_0 = N - I_0
R_0 = 0.
delta = .1

ts = np.arange(0, 20, delta)
S, I, R = S_0, I_0, R_0
Ss, Is, Rs = [], [], []
beta = 2.3
gamma = .07
for t in ts:
    S_next = S - beta*S*I*delta/N
    I_next = I + beta*S*I*delta/N - gamma*I
    R_next = R + gamma*I
    S, I, R = S_next, I_next, R_next
    Ss.append(S)
    Is.append(I)
    Rs.append(R)

fig, ax = plt.subplots()
data = np.vstack([Ss, Is, Rs])
ax.stackplot(ts, data, labels=["S", "I", "R"])
#ax.plot(ts, Ss)
#ax.plot(ts, Is)
#ax.plot(ts, Rs)
ax.legend();
```

### &beta; exploration

```python
bs = np.arange(0, 3, .1)
Ss, Is, Rs, mxIs = [], [], [], []
for beta in bs:
    N = 800.
    I_0 = 1.
    S_0 = N - I_0
    R_0 = 0.
    delta = .1
    ts = np.arange(0, 100, delta)
    S, I, R = S_0, I_0, R_0
    max_I = 0
    gamma = .07
    for t in ts:
        S_next = S - beta*S*I*delta/N
        I_next = I + beta*S*I*delta/N - gamma*I
        R_next = R + gamma*I
        S, I, R = S_next, I_next, R_next
        if I > max_I:
            max_I = I
    Ss.append(S)
    Is.append(I)
    Rs.append(R)
    mxIs.append(mxIs)

fig, ax = plt.subplots()
ax.plot(bs, Ss, label="S")
ax.plot(bs, Is, label="I")
ax.plot(bs, Rs, label="R")
ax.plot(bs, Rs, label="mxIs")
ax.legend();
```

### Thougts
Increasing &beta; decreases number of people not affected by the disease. This means that for the general population the lower the &beta; parameter, the better. Worth noting that the lowest number of infected at the same time is when &beta; is smalles. #flattenthecurve

### &gamma; exploration

```python
ys = np.arange(0, 1, .001)
Ss, Is, Rs, mxIs = [], [], [], []
for gamma in ys:
    N = 800.
    I_0 = 1.
    S_0 = N - I_0
    R_0 = 0.
    delta = .1
    ts = np.arange(0, 100, delta)
    S, I, R = S_0, I_0, R_0
    max_I = 0
    beta = 1.5
    for t in ts:
        S_next = S - beta*S*I*delta/N
        I_next = I + beta*S*I*delta/N - gamma*I
        R_next = R + gamma*I
        S, I, R = S_next, I_next, R_next
        if I > max_I:
            max_I = I
    Ss.append(S)
    Is.append(I)
    Rs.append(R)
    mxIs.append(mxIs)

fig, ax = plt.subplots()
ax.plot(ys, Ss, label="S")
ax.plot(ys, Is, label="I")
ax.plot(ys, Rs, label="R")
ax.plot(ys, Rs, label="mxIs")
ax.legend();
```

### Thoughts
Higher &gamma;s aren't necesarily worse - this is due to the fact that infected don't manage to spread the disease before being removed from the infected pool.
