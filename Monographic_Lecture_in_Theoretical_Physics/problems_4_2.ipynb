```python
import numpy as np
import matplotlib.pyplot as plt
from time import time
```

### 4.2) Square root using Newton's method
Newton's method can be formulated as:
$$
x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}
$$
Looking for a square root of a given $a$ will mean that we're looking for $x$ such that $x^2=a$. This in turn means we're looking for $f(x)$ such that $f(x)=x^2-a$. So our formula will look as follows:
$$
x_{n+1} = x_n - \frac{x_n^2-a}{2x_n}
$$
Let's define our parameters:

```python
a = time()
x_0 = a/2
eps = .1
```

Function definition

```python
def next_approx(x):
    return x - ((x**2)-a)/(2*x)
```

Calculations

```python
x = x_0
err = abs(np.sqrt(a)-x)
xs = [x]
errs = [err]
while (err > eps):
    x = next_approx(x)
    err = abs(np.sqrt(a)-x)
    xs.append(x)
    errs.append(err)
```

```python
plt.figure(figsize=(12,6))
plt.title(f"Approximation of square root of {a}", fontsize=20)
plt.plot(errs, label="Error")
plt.plot(xs, label="Approximate value")
plt.legend()
plt.yscale("log");
```

Newton's method can fail to converge when:
1. If the root approaches 0 as x approaches infinity - since x will keep on increasing. In such case another method should be used.
1. If the start from a stationary point, the $f'(x_n)$, which is in denominator, becomes $0$.
1. Since we're taking a previous value as an input to a next one a function that changes it's sign can cause a loop.
1. We're using derivative which means our function needs to be differentiable on range that can't be exceeded during iterations
