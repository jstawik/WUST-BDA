```python
import numpy as np
import requests
import io
from PIL import Image
from IPython.display import display 

query = "tell me a joke"

appid = open("app_id.secret", "r").read()
with open("app_id.secret") as secret:
    appid = secret.read()
url = "https://api.wolframalpha.com/v1/simple?"
params = {"appid": appid.rstrip("\n"), "units": "metric", "i": query, "background":"transparent", "foreground":"black"}
r = requests.get(url, params)

image_data = r.content
image = Image.open(io.BytesIO(image_data))
display(image)
```
