### Fixed Point Iteration 1 

Let f(x)=x<sup>2</sup>+x−5

1. Prove (in the notebook) that f(x) has a root in [1,2]
2. Find a function g(x) that could be used as a fixed point iteration formula to find a root of f(x) on [1,2]. Assume x<sub>0</sub>=1.5. Write a program which calculates the root with accuracy of the order of $\epsilon$ =10<sup>−4</sup>.

Ad.1.: Roots of f(x) are: $\frac{-1\pm\sqrt{21}}{2}$.\
$1 < \frac{-1+\sqrt{21}}{2} < 2$

Ad.2.: The equation for fixed point iteration method looks as follows:
$$f(x) = x^2 + x -5 \Longrightarrow x = -1 + \frac{5}{x} \Longrightarrow g(x) = -1 + \frac{5}{x}$$


```python
x_init = 1.5
eps = 10e-4


def next_approx(x):
    return -1 + 5/x
root = (-1 + np.sqrt(21))/2

x = x_init
err = abs(root - x_init)
xs = [x_init]
errs = [err]
for i in range(30):
    if err < eps:
        break
    x = next_approx(x)
    err = abs(root - x)
    xs.append(x)
    errs.append(err)

plt.figure(figsize=(12,6))
plt.title(f"Approximation root", fontsize=20)
plt.plot(errs, label="Error")
plt.plot(xs, label="Approximate value")
plt.legend();
if err > eps:
    plt.text(24, 5, "Won't converge");
```

For basic parameters the above failed. Let's try:
$$g(x) = \sqrt{5-x}$$

```python
def next_approx(x):
    return np.sqrt(5-x)
root = (-1 + np.sqrt(21))/2

x = x_init
err = abs(root - x_init)
xs = [x_init]
errs = [err]
for i in range(300):
    if err < eps:
        break
    x = next_approx(x)
    err = abs(root - x)
    xs.append(x)
    errs.append(err)

plt.figure(figsize=(12,6))
plt.title(f"Approximation root", fontsize=20)
plt.plot(errs, label="Error")
plt.plot(xs, label="Approximate value")
plt.legend();
```
