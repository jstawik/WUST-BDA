### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Mass-on-spring oscillator with damping and driven by an external force

```python
param_k = .5
param_m = 4
param_F0 = .8
param_w = .1
init_x, init_v = 1, 0

delta = .1

ts = np.arange(0, 20*np.pi*np.sqrt(param_m/param_k), delta)
force = param_F0*np.sin(param_w*ts)

xs, vs = [], []

x, v = init_x, init_v
for t in range(len(ts)):
    v = v + delta*force[t] - delta*param_k*x
    x = x + v*delta
    vs.append(v)
    xs.append(x)



plt.plot(ts, xs)
plt.plot(ts, force);
```
