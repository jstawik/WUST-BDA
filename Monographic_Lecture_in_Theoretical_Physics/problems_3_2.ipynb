### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Free fall with nonzero initial velocity
Parameters setup

```python
const_g = 9.81
delta = .1
ts = np.arange(0,7,delta)
init_h = 100
init_v = -10
```

Euler method

```python
h = init_h
v = init_v

euler_traj = {"height":[], "velocity":[]}
for t in ts:
    h = h - v * delta
    v = v + const_g*delta
    if h > 0:
        euler_traj["height"].append(h)
        euler_traj["velocity"].append(v)
    else:
        euler_traj["height"].append(0)
        euler_traj["velocity"].append(0)
        
plt.figure(figsize=(12,5))        
plt.title("Euler")
plt.scatter(ts, euler_traj["height"], marker='.', label="Height")
plt.scatter(ts, euler_traj["velocity"], marker='.', label="Velocity")
plt.legend()
plt.show()
```

Midpoint method

```python
h = init_h
v = init_v

midpoint_traj = {"height":[], "velocity":[]}
for t in ts:
    kv = (v + v + const_g*delta)/2
    kv2 = (kv + kv + const_g*delta)/2
    
    kh = (h + h - v*delta)/2
    kh2 = (kh + kh - v* delta)/2
    
    v = kv2
    h = kh2
    if h > 0:
        midpoint_traj["height"].append(h)
        midpoint_traj["velocity"].append(v)
    else:
        midpoint_traj["height"].append(0)
        midpoint_traj["velocity"].append(0)
        
plt.figure(figsize=(12,5))        
plt.title("Midpoint")
plt.scatter(ts, midpoint_traj["height"], marker='.', label="Height")
plt.scatter(ts, midpoint_traj["velocity"], marker='.', label="Velocity")
plt.legend()
plt.show()
```

4<sup>th</sup> order Runge–Kutta  method

```python
h = init_h
v = init_v

runge_traj = {"height":[], "velocity":[]}
for t in ts:
    k1h = init_v + const_g*t
    k2h = init_v + const_g*(t+delta/2)
    k3h = init_v + const_g*(t+delta/2) 
    k4h = init_v + const_g*(t+delta)
    
    k1v = const_g
    k2v = const_g
    k3v = const_g
    k4v = const_g
    
    h = h - delta*(k1h + 2*k2h + 2*k3h + k4h)/6
    v = v + delta*(k1v + 2*k2v + 2*k3v + k4v)/6
    if h > 0:
        runge_traj["height"].append(h)
        runge_traj["velocity"].append(v)
    else:
        runge_traj["height"].append(0)
        runge_traj["velocity"].append(0)
  
plt.figure(figsize=(12,5))        
plt.title("Runge–Kutta")
plt.scatter(ts, runge_traj["height"], marker='.', label="Height")
plt.scatter(ts, runge_traj["velocity"], marker='.', label="Velocity");
plt.legend()
plt.show()
```

Comparison

```python
plt.figure(figsize=(12,5))        
plt.title("Height")
plt.scatter(ts, euler_traj["height"], marker='.', label="Euler")
plt.scatter(ts, midpoint_traj["height"], marker='.', label="Midpoint")
plt.scatter(ts, runge_traj["height"], marker='.', label="Runge–Kutta")
plt.legend()
plt.show()
plt.figure(figsize=(12,5))        
plt.title("Velocity")
plt.scatter(ts, euler_traj["velocity"], marker='.', label="Euler")
plt.scatter(ts, midpoint_traj["velocity"], marker='.', label="Midpoint")
plt.scatter(ts, runge_traj["velocity"], marker='.', label="Runge–Kutta")
plt.legend();
```
