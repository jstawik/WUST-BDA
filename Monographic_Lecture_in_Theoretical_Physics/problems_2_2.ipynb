```python
import numpy as np
import matplotlib.pyplot as plt
```

### Predator-Prey model
$$
\begin{cases}
\frac{dr(t)}{dt} = 2r(t)-\alpha r(t)f(t) \\
\frac{df(t)}{dt} = -f+\alpha r(t)f(t)
\end{cases}
$$

```python
r_0 = 300
f_0 = 150
a = .01
delta = .001

ts = np.arange(0, 70, delta)
r, f = r_0, f_0
rs, fs = [], []

for t in ts:
    r_next = r + (2*r-a*r*f)*delta
    f_next = f + (-f + a*r*f)*delta
    r, f = max(r_next, 0), max(f_next, 0)
    rs.append(r)
    fs.append(f)

fig, ax = plt.subplots()
data = np.vstack([rs, fs])
ax.plot(ts, rs)
ax.plot(ts, fs);
```
